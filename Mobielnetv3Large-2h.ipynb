```python
import os
import glob
import numpy as np
import pandas as pd
import tensorflow as tf
import pickle
import matplotlib.pyplot as plt
import seaborn as sns

from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.applications import MobileNetV3Large
from tensorflow.keras.optimizers import AdamW # <--- MEJORA: Usando el optimizador AdamW, más estable
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from sklearn.metrics import classification_report, confusion_matrix

# --- INICIO DE FUNCIONES AUXILIARES (Sin cambios) ---
def get_label_from_pts(pts_path):
    try:
        with open(pts_path, 'r') as f:
            lines = f.readlines()
            for line in lines:
                if line.strip().startswith('label:'):
                    return int(line.strip().split(':')[1].strip())
    except FileNotFoundError:
        return None
    return None

def create_dataframe_for_flow(directory):
    filepaths = []
    labels = []
    image_paths = glob.glob(os.path.join(directory, '*.jpg'))
    if not image_paths:
        print(f"Advertencia: No se encontraron archivos .jpg en la ruta: {directory}")
        return pd.DataFrame({'filepaths': [], 'labels': []})
    for img_path in image_paths:
        pts_path = img_path.replace('.jpg', '.pts')
        label = get_label_from_pts(pts_path)
        if label is not None:
            filepaths.append(img_path)
            labels.append('despierto' if label == 0 else 'somnoliento')
        else:
            print(f"Advertencia: No se encontró el archivo .pts para la imagen: {os.path.basename(img_path)}")
    return pd.DataFrame({'filepaths': filepaths, 'labels': labels})
# --- FIN DE FUNCIONES AUXILIARES ---

if __name__ == "__main__":
    # --- Configuración (Sin cambios) ---
    base_dir = '/content/drive/MyDrive/Dataset/Dataset 29'
    train_dir = os.path.join(base_dir, 'train')
    validation_dir = os.path.join(base_dir, 'validation')
    test_dir = os.path.join(base_dir, 'test')

    image_size = (224, 224)
    batch_size = 32
    epochs = 50

    # --- Carga y preprocesamiento de datos (Sin cambios) ---
    print("Cargando y procesando los datos...")
    train_df = create_dataframe_for_flow(train_dir)
    validation_df = create_dataframe_for_flow(validation_dir)
    test_df = create_dataframe_for_flow(test_dir)

    if train_df.empty or validation_df.empty or test_df.empty:
        print("\nError: No se pudieron cargar los datos.")
    else:
        # (Generadores de datos sin cambios)
        train_datagen = ImageDataGenerator(
            rescale=1./255., rotation_range=20, width_shift_range=0.2,
            height_shift_range=0.2, shear_range=0.2, zoom_range=0.2,
            horizontal_flip=True, fill_mode='nearest'
        )
        validation_test_datagen = ImageDataGenerator(rescale=1./255.)
        train_generator = train_datagen.flow_from_dataframe(
            dataframe=train_df, x_col='filepaths', y_col='labels',
            target_size=image_size, batch_size=batch_size, class_mode='binary', shuffle=True
        )
        validation_generator = validation_test_datagen.flow_from_dataframe(
            dataframe=validation_df, x_col='filepaths', y_col='labels',
            target_size=image_size, batch_size=batch_size, class_mode='binary', shuffle=False
        )
        test_generator = validation_test_datagen.flow_from_dataframe(
            dataframe=test_df, x_col='filepaths', y_col='labels',
            target_size=image_size, batch_size=batch_size, class_mode='binary', shuffle=False
        )

        # --- Construcción del modelo simplificado ---
        print("\nCargando modelo y construyendo clasificador simplificado...")
        base_model = MobileNetV3Large(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

        x = base_model.output
        x = GlobalAveragePooling2D()(x)
        # <--- MEJORA: Clasificador aún más simple para mayor estabilidad ---
        x = Dense(256, activation='relu')(x)
        x = Dropout(0.5)(x) # Volvemos a 0.5, un valor estándar y robusto
        predictions = Dense(1, activation='sigmoid')(x)
        model = Model(inputs=base_model.input, outputs=predictions)

        # <--- ESTRATEGIA SIMPLIFICADA: Fine-tuning conservador y en una sola fase ---
        base_model.trainable = True

        # Descongelamos solo las últimas 15 capas. Muy conservador para buscar estabilidad.
        for layer in base_model.layers[:-15]:
            layer.trainable = False

        # Compilamos con AdamW y una tasa de aprendizaje moderada y única.
        model.compile(optimizer=AdamW(learning_rate=1e-4),
                      loss='binary_crossentropy',
                      metrics=['accuracy'])

        model.summary()

        # --- Callbacks ---
        callbacks = [
            EarlyStopping(monitor='val_loss', patience=10, verbose=1, restore_best_weights=True), # Aumentamos la paciencia
            ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=4, verbose=1, min_lr=1e-6),
            ModelCheckpoint('modelo_somnolencia_mobilenetv3_stable.h5', monitor='val_loss', save_best_only=True)
        ]

        # --- Entrenamiento de Fase Única ---
        print("\nIniciando entrenamiento simplificado para buscar estabilidad...")
        history = model.fit(
            train_generator,
            steps_per_epoch=train_generator.n // train_generator.batch_size,
            validation_data=validation_generator,
            validation_steps=validation_generator.n // validation_generator.batch_size,
            epochs=epochs,
            callbacks=callbacks
            # Nota: Hemos quitado class_weight intencionadamente
        )

        # --- Guardar historial ---
        print("\nGuardando el historial de entrenamiento...")
        with open('training_history_mobilenetv3_stable.pkl', 'wb') as file:
            pickle.dump(history.history, file)
        print("Historial guardado en 'training_history_mobilenetv3_stable.pkl'")

        # --- Evaluación del modelo ---
        print("\nEvaluando el mejor modelo en el conjunto de prueba...")
        test_loss, test_accuracy = model.evaluate(test_generator)
        print(f"\nPrecisión en el conjunto de prueba: {test_accuracy:.4f}")

        # (El resto del código de evaluación es el mismo)
        test_generator.reset()
        y_pred_probs = model.predict(test_generator)
        y_pred_classes = (y_pred_probs > 0.5).astype(int)
        class_indices = test_generator.class_indices
        labels_list = list(class_indices.keys())

        print("\nReporte de Clasificación:")
        print(classification_report(test_generator.classes, y_pred_classes, target_names=labels_list))

        conf_matrix = confusion_matrix(test_generator.classes, y_pred_classes)
        print("\nMatriz de Confusión:")
        print(conf_matrix)

        plt.figure(figsize=(8, 6))
        sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=labels_list, yticklabels=labels_list)
        plt.xlabel('Predicción')
        plt.ylabel('Etiqueta Real')
        plt.title('Matriz de Confusión')
        plt.show()
```
