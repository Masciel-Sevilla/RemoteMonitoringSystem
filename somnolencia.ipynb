```python
# Celda 2: Código completo para entrenar, guardar y evaluar el modelo

import os
import glob
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Función para leer las etiquetas desde los archivos .pts
def get_label_from_pts(pts_path):
    """
    Lee un archivo .pts y extrae la etiqueta de clasificación.
    Devuelve la etiqueta (0 o 1) si se encuentra, de lo contrario None.
    """
    try:
        with open(pts_path, 'r') as f:
            lines = f.readlines()
            for line in lines:
                if line.strip().startswith('label:'):
                    return int(line.strip().split(':')[1].strip())
    except FileNotFoundError:
        return None
    return None

# Función para crear un DataFrame con las rutas de las imágenes y sus etiquetas
def create_dataframe_for_flow(directory):
    """
    Crea un DataFrame de Pandas con las rutas de los archivos .jpg y sus etiquetas.
    Las etiquetas se obtienen de los archivos .pts correspondientes.
    """
    filepaths = []
    labels = []

    image_paths = glob.glob(os.path.join(directory, '*.jpg'))

    if not image_paths:
        print(f"Advertencia: No se encontraron archivos .jpg en la ruta: {directory}")
        return pd.DataFrame({'filepaths': [], 'labels': []})

    for img_path in image_paths:
        pts_path = img_path.replace('.jpg', '.pts')
        label = get_label_from_pts(pts_path)

        if label is not None:
            filepaths.append(img_path)
            labels.append('despierto' if label == 0 else 'somnoliento')
        else:
            print(f"Advertencia: No se encontró el archivo .pts para la imagen: {os.path.basename(img_path)}")

    return pd.DataFrame({'filepaths': filepaths, 'labels': labels})

# El bloque principal para la ejecución del script
if __name__ == "__main__":

    # ------------------
    # Configuración de las rutas y parámetros
    # ------------------

      # Definir la ruta a la carpeta principal donde están los archivos .jpg y .pts
    base_dir = '/content/drive/MyDrive/Dataset/Dataset 29'



    # Estas líneas son las correctas y asumen que las subcarpetas existen.
    train_dir = os.path.join(base_dir, 'train')
    validation_dir = os.path.join(base_dir, 'validation')
    test_dir = os.path.join(base_dir, 'test')

    image_size = (224, 224)
    batch_size = 32
    epochs = 10 # Puedes ajustar el número de épocas según sea necesario

    # ------------------
    # Carga y preprocesamiento de datos
    # ------------------

    print("Cargando y procesando los datos...")
    train_df = create_dataframe_for_flow(train_dir)
    validation_df = create_dataframe_for_flow(validation_dir)
    test_df = create_dataframe_for_flow(test_dir)

    if train_df.empty or validation_df.empty or test_df.empty:
        print("\nError: No se pudieron cargar los datos correctamente. Por favor, verifica la ruta y la existencia de los archivos .jpg y .pts.")
    else:
        print(f"\nDatos de entrenamiento: {len(train_df)} muestras")
        print(f"Datos de validación: {len(validation_df)} muestras")
        print(f"Datos de prueba: {len(test_df)} muestras")

        # Configuración del generador de imágenes
        datagen = ImageDataGenerator(rescale=1./255.) # Normalizar las imágenes

        # Creación de los generadores de datos
        train_generator = datagen.flow_from_dataframe(
            dataframe=train_df,
            x_col='filepaths',
            y_col='labels',
            target_size=image_size,
            batch_size=batch_size,
            class_mode='binary',
            shuffle=True
        )

        validation_generator = datagen.flow_from_dataframe(
            dataframe=validation_df,
            x_col='filepaths',
            y_col='labels',
            target_size=image_size,
            batch_size=batch_size,
            class_mode='binary',
            shuffle=False
        )

        test_generator = datagen.flow_from_dataframe(
            dataframe=test_df,
            x_col='filepaths',
            y_col='labels',
            target_size=image_size,
            batch_size=batch_size,
            class_mode='binary',
            shuffle=False
        )

        # ------------------
        # Construcción y compilación del modelo
        # ------------------

        print("\nCargando el modelo ResNet-50 pre-entrenado...")
        base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

        # Congelar las capas convolucionales de ResNet-50
        for layer in base_model.layers:
            layer.trainable = False

        # Agregar una capa de clasificación personalizada
        x = base_model.output
        x = GlobalAveragePooling2D()(x)
        x = Dense(1024, activation='relu')(x)
        predictions = Dense(1, activation='sigmoid')(x)

        model = Model(inputs=base_model.input, outputs=predictions)

        model.compile(optimizer=Adam(learning_rate=0.0001),
                      loss='binary_crossentropy',
                      metrics=['accuracy'])

        print("\nResumen del modelo:")
        model.summary()

        # ------------------
        # Entrenamiento del modelo
        # ------------------

        print("\nIniciando el entrenamiento...")
        history = model.fit(
            train_generator,
            steps_per_epoch=train_generator.n // train_generator.batch_size,
            validation_data=validation_generator,
            validation_steps=validation_generator.n // validation_generator.batch_size,
            epochs=epochs
        )

        # ------------------
        # Guardar el modelo
        # ------------------

        model_save_path = 'modelo_somnolencia_resnet50.h5'
        model.save(model_save_path)
        print(f"\nModelo entrenado guardado en: {model_save_path}")

        # ------------------
        # Evaluación del modelo
        # ------------------

        print("\nEvaluando el modelo en el conjunto de prueba...")
        test_loss, test_accuracy = model.evaluate(test_generator)
        print(f"\nPrecisión en el conjunto de prueba: {test_accuracy:.4f}")

        test_generator.reset()
        y_pred_probs = model.predict(test_generator)
        y_pred_classes = (y_pred_probs > 0.5).astype(int)

        class_indices = test_generator.class_indices
        labels_list = list(class_indices.keys())

        print("\nReporte de Clasificación:")
        print(classification_report(test_generator.classes, y_pred_classes, target_names=labels_list))

        conf_matrix = confusion_matrix(test_generator.classes, y_pred_classes)
        print("\nMatriz de Confusión:")
        print(conf_matrix)

        plt.figure(figsize=(8, 6))
        sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=labels_list, yticklabels=labels_list)
        plt.xlabel('Predicción')
        plt.ylabel('Etiqueta Real')
        plt.title('Matriz de Confusión')
        plt.show()
```
