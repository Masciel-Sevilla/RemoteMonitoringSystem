```python
import os
import glob
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# --- INICIO DE FUNCIONES AUXILIARES (CORRECCIÓN) ---

# Función para leer las etiquetas desde los archivos .pts
def get_label_from_pts(pts_path):
    """
    Lee un archivo .pts y extrae la etiqueta de clasificación.
    Devuelve la etiqueta (0 o 1) si se encuentra, de lo contrario None.
    """
    try:
        with open(pts_path, 'r') as f:
            lines = f.readlines()
            for line in lines:
                if line.strip().startswith('label:'):
                    return int(line.strip().split(':')[1].strip())
    except FileNotFoundError:
        return None
    return None

# Función para crear un DataFrame con las rutas de las imágenes y sus etiquetas
def create_dataframe_for_flow(directory):
    """
    Crea un DataFrame de Pandas con las rutas de los archivos .jpg y sus etiquetas.
    Las etiquetas se obtienen de los archivos .pts correspondientes.
    """
    filepaths = []
    labels = []

    image_paths = glob.glob(os.path.join(directory, '*.jpg'))

    if not image_paths:
        print(f"Advertencia: No se encontraron archivos .jpg en la ruta: {directory}")
        return pd.DataFrame({'filepaths': [], 'labels': []})

    for img_path in image_paths:
        pts_path = img_path.replace('.jpg', '.pts')
        label = get_label_from_pts(pts_path)

        if label is not None:
            filepaths.append(img_path)
            labels.append('despierto' if label == 0 else 'somnoliento')
        else:
            print(f"Advertencia: No se encontró el archivo .pts para la imagen: {os.path.basename(img_path)}")

    return pd.DataFrame({'filepaths': filepaths, 'labels': labels})

# --- FIN DE FUNCIONES AUXILIARES ---


# El bloque principal para la ejecución del script
if __name__ == "__main__":

    # ------------------
    # Configuración de las rutas y parámetros
    # ------------------
    base_dir = '/content/drive/MyDrive/Dataset/Dataset 29'
    train_dir = os.path.join(base_dir, 'train')
    validation_dir = os.path.join(base_dir, 'validation')
    test_dir = os.path.join(base_dir, 'test')

    image_size = (224, 224)
    batch_size = 32
    epochs = 50  # Aumentamos las épocas; EarlyStopping decidirá cuándo parar

    # ------------------
    # Carga y preprocesamiento de datos
    # ------------------
    print("Cargando y procesando los datos...")
    train_df = create_dataframe_for_flow(train_dir)
    validation_df = create_dataframe_for_flow(validation_dir)
    test_df = create_dataframe_for_flow(test_dir)

    if train_df.empty or validation_df.empty or test_df.empty:
        print("\nError: No se pudieron cargar los datos.")
    else:
        print(f"\nDatos de entrenamiento: {len(train_df)} muestras")
        print(f"Datos de validación: {len(validation_df)} muestras")
        print(f"Datos de prueba: {len(test_df)} muestras")

        # MEJORA 1: Aumento de datos para el conjunto de entrenamiento
        train_datagen = ImageDataGenerator(
            rescale=1./255.,
            rotation_range=20,
            width_shift_range=0.2,
            height_shift_range=0.2,
            shear_range=0.2,
            zoom_range=0.2,
            horizontal_flip=True,
            fill_mode='nearest'
        )

        validation_test_datagen = ImageDataGenerator(rescale=1./255.)

        train_generator = train_datagen.flow_from_dataframe(
            dataframe=train_df, x_col='filepaths', y_col='labels',
            target_size=image_size, batch_size=batch_size, class_mode='binary', shuffle=True
        )
        validation_generator = validation_test_datagen.flow_from_dataframe(
            dataframe=validation_df, x_col='filepaths', y_col='labels',
            target_size=image_size, batch_size=batch_size, class_mode='binary', shuffle=False
        )
        test_generator = validation_test_datagen.flow_from_dataframe(
            dataframe=test_df, x_col='filepaths', y_col='labels',
            target_size=image_size, batch_size=batch_size, class_mode='binary', shuffle=False
        )

        # ------------------
        # Construcción y compilación del modelo
        # ------------------
        print("\nCargando el modelo ResNet-50 pre-entrenado...")
        base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

        # MEJORA 2: Se agrega capa de Dropout para regularización
        x = base_model.output
        x = GlobalAveragePooling2D()(x)
        x = Dense(1024, activation='relu')(x)
        x = Dropout(0.5)(x)
        predictions = Dense(1, activation='sigmoid')(x)

        model = Model(inputs=base_model.input, outputs=predictions)

        # MEJORA 3: Fine-Tuning (descongelar capas superiores)
        for layer in base_model.layers[:143]:
            layer.trainable = False
        for layer in base_model.layers[143:]:
            layer.trainable = True

        # Compilamos con una tasa de aprendizaje más baja para el fine-tuning
        model.compile(optimizer=Adam(learning_rate=1e-5), # 0.00001
                      loss='binary_crossentropy',
                      metrics=['accuracy'])

        print("\nResumen del modelo mejorado:")
        model.summary()

        # ------------------
        # Entrenamiento del modelo
        # ------------------
        # MEJORA 4: Uso de Callbacks para un entrenamiento más inteligente
        callbacks = [
            EarlyStopping(monitor='val_loss', patience=5, verbose=1, restore_best_weights=True),
            ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, verbose=1, min_lr=1e-6),
            ModelCheckpoint('modelo_somnolencia_resnet50_best.h5', monitor='val_loss', save_best_only=True)
        ]

        print("\nIniciando el entrenamiento mejorado...")
        history = model.fit(
            train_generator,
            steps_per_epoch=train_generator.n // train_generator.batch_size,
            validation_data=validation_generator,
            validation_steps=validation_generator.n // validation_generator.batch_size,
            epochs=epochs,
            callbacks=callbacks
        )

        # ------------------
        # Evaluación del modelo
        # ------------------
        print("\nEvaluando el mejor modelo en el conjunto de prueba...")
        test_loss, test_accuracy = model.evaluate(test_generator)
        print(f"\nPrecisión en el conjunto de prueba: {test_accuracy:.4f}")

        test_generator.reset()
        y_pred_probs = model.predict(test_generator)
        y_pred_classes = (y_pred_probs > 0.5).astype(int)

        class_indices = test_generator.class_indices
        labels_list = list(class_indices.keys())

        print("\nReporte de Clasificación:")
        print(classification_report(test_generator.classes, y_pred_classes, target_names=labels_list))

        conf_matrix = confusion_matrix(test_generator.classes, y_pred_classes)
        print("\nMatriz de Confusión:")
        print(conf_matrix)

        plt.figure(figsize=(8, 6))
        sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=labels_list, yticklabels=labels_list)
        plt.xlabel('Predicción')
        plt.ylabel('Etiqueta Real')
        plt.title('Matriz de Confusión')
        plt.show()
```
